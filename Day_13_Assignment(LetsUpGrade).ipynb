```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style("darkgrid")
```

```python

plt.plot(creditcard_df.Amount,creditcard_df.Time)
#plt.plot(creditcard_df.Amount,creditcard_df.V2, 'o--r')

plt.xlabel('Amount')
plt.ylabel('Time')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot( creditcard_df.Time,
                 creditcard_df.Amount,
                 hue=creditcard_df.Class,  
                 s=100);

plt.title("Flowers");
```
